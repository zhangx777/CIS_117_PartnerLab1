```
# Combined Games
# Group 6
# Tori Zhang and Geoffrey Won
# 09-24-2025

import random

def play_guessing_game():
    """
    Implements a number guessing game.
    The function generates a random number between 1 and 100, gives the user
    5 tries to guess it, and provides hints along the way.
    The user is prompted to play again after each game.
    """
    play_again = 'Y'

    # The main game loop that allows the user to play multiple times.
    while play_again.upper() == 'Y':
        # Generate a random number between 1 and 100 (inclusive).
        secret_number = random.randint(1, 100)

        # Set the number of tries the user gets.
        tries = 5

        print("I'm thinking of a number between 1 and 100.")
        print(f"Guess what it is. You have {tries} tries:")

        # The inner loop for a single game session.
        while tries > 0:
            try:
                # Prompt the user for a guess.
                guess = int(input())

                # Check if the guess is correct.
                if guess == secret_number:
                    print("You got it!")
                    # Exit the inner loop if the guess is correct.
                    break

                # Provide hints based on the guess.
                elif guess < secret_number:
                    print("Nope! Too low. ", end="")
                else:
                    print("Nope! Too high. ", end="")

                # Decrement the number of tries.
                tries -= 1

                # Inform the user of their remaining tries.
                if tries > 0:
                    print(f"Try again ({tries} tries left): ", end="")
                else:
                    print(f"You lost. The number was {secret_number}")

            except ValueError:
                # Handle cases where the user enters non-integer input.
                print("Invalid input. Please enter a number.")

        # Ask the user if they want to play again.
        play_again = input("Do you want to play again? (Y/N): ")

# start of the rock paper scissors game
def play_rps_game():
    choice = input("Do you want to play? (yes/no): ")
    if (choice.lower() == "yes"):
        while choice.lower() == "yes":

            # Computer and player choices prompted
            # Computer as the instructions said is a random int
            computer = random.randint(1, 3)
            player = int(input("Player, please enter 1: rock, 2: paper, or 3: scissors: "))

            # so this is called a "dictionary" i had to watch a video on how this works. I dont think we learned it yet, please dont take off points :)
            choices = {1: "rock", 2: "paper", 3: "scissors"}

            # dictionary basically maps an key which is the label (in this case an int) and a value (rock paper scissors)
            # so the player and the computer's choices correspond to a number as the instructions stated
            player_choice = choices[player]
            computer_choice = choices[computer]

            # prints the computer and the players choice for clarity
            print(f"\nYou chose: {player_choice}")
            print(f"Computer chose: {computer_choice}")

            # if player and computer's choice is the same it's a tie
            if player_choice == computer_choice:
                    print("It's a tie!\n")

            # to avoid repetition i used the "or" statement which i also had to watch a video on how to use. It eliminates the repetition from using elif over and over
            elif (player_choice == "rock" and computer_choice == "scissors") \
                or (player_choice == "paper" and computer_choice == "rock") \
                or (player_choice == "scissors" and computer_choice == "paper"):

                print("Player Wins!\n")
            else:
                    print("Computer Wins!\n")

            choice = input("Would you like to play again? (yes/no): ").lower()

    # if your initial choice was no then the game prints a farewell
    if (choice.lower() == "no"):
        print("See you next time!\n")


if __name__ == "__main__":
    while True:
        print("Which game do you want to play?")
        print("1. Guessing Game")
        print("2. Rock-paper-scissors")
        print("3. Exit")

        game_choice = input("Enter your choice (1, 2, or 3): ")

        if game_choice == '1':
            play_guessing_game()
        elif game_choice == '2':
            play_rps_game()
        elif game_choice == '3':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")
```
